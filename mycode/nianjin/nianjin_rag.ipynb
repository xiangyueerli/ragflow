```python
def retrieve_tests(question, ds=dataset):
    # rag = RAGFlow(API_KEY, HOST_ADDRESS)
    rag = rag_object
    chunks = rag.retrieve(dataset_ids=[ds.id], top_k=3, similarity_threshold=0.0,vector_similarity_weight=1,question=question,document_ids=[],page=1,page_size=10)
    return chunks
chunks = retrieve_tests(question=questions[0])
print(chunks)
```

```python
dict_ = {}
for idx, question in enumerate([questions[8]]):
    chunks = retrieve_tests(question, ds=dataset)
    content = [chunk.content for chunk in chunks]
    
    dict_[idx+1] = content
dict_[1]
```
